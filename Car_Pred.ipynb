```python
#Dropping Columns that will not be needed for the prediction
Car_df.drop(columns=['model','color','vin', 'lot', 'state',], axis=1, inplace=True)
Car_df
```

```python
# Convert the 'year' column back to integer for the age calculation
Car_df['year'] = Car_df['year'].astype(int)
Car_df.info
```

```python
# Changing the year column to a vehicle age column
current_year = datetime.datetime.now().year
Car_df['vehicle_age'] = current_year-Car_df['year']
Car_df
```

```python
##Droping the year since we are going to use the vehicle age instead
Car_df.drop('year', axis =1, inplace=True)
Car_df
```

```python
# Function to extract hours from "condition" column
def extract_hours(condition_string):
    if 'days' in condition_string:
        days = int(condition_string.split()[0])
        return days * 24
    elif 'hours' in condition_string:
        return int(condition_string.split()[0])
    else:
        return 0
```

```python
# Applying the function to the condition to create a new column called hours_left
Car_df['hours_left'] = Car_df['condition'].apply(extract_hours)
Car_df
```

```python
#Dropping the condition column, since we would use the hours_left column
Car_df.drop('condition', axis=1, inplace=True)
Car_df
```
